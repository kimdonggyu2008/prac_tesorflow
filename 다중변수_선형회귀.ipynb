```python
import tensorflow as tf
import numpy as np

print(tf.__version__)
```

1차원 데이터 2개


```python
from logging import basicConfig
x1_data = [1, 0, 3, 0, 5]
x2_data = [0, 2, 0, 4, 0]
y_data  = [1, 2, 3, 4, 5]

W1=tf.Variable(tf.random.uniform((1,),-10.0,10.0))
W2 = tf.Variable(tf.random.uniform((1,), -10.0, 10.0))
b  = tf.Variable(tf.random.uniform((1,), -10.0, 10.0))

learning_rate=tf.Variable(0.001)

for i in range(1000+1):
  with tf.GradientTape() as tape:
    hypothesis=W1*x1_data+W2*x2_data+b#현재 예측값
    cost=tf.reduce_mean(tf.square(hypothesis-y_data))#예측값과 실제값 절댓값바꾸기
  W1_grad,W2_grad,b_grad=tape.gradient(cost,[W1,W2,b])#기울기 계산하기
  W1.assign_sub(learning_rate*W1_grad)#업데이트
  W2.assign_sub(learning_rate*W2_grad)
  b.assign_sub(learning_rate*b_grad)

  if i % 50 == 0:
    print("{:5} | {:10.6f} | {:10.4f} | {:10.4f} | {:10.6f}".format(
      i, cost.numpy(), W1.numpy()[0], W2.numpy()[0], b.numpy()[0]))
```

2차원 배열


```python
x_data = [
    [1., 0., 3., 0., 5.],
    [0., 2., 0., 4., 0.]
]
y_data  = [1, 2, 3, 4, 5]

W=tf.Variable(tf.random.uniform((1,2),-1.0,1.0))
b=tf.Variable(tf.random.uniform((1,),-1.0,1.0))
learning_rate = tf.Variable(0.001)

for i in range(1000+1):
  with tf.GradientTape() as tape:
    hypothesis=tf.matmul(W,x_data)+b #스칼라 배열 계산불가
    cost=tf.reduce_mean(tf.square(hypothesis-y_data))#계산

    W_grad, b_grad = tape.gradient(cost, [W, b])#적용
    W.assign_sub(learning_rate * W_grad)
    b.assign_sub(learning_rate * b_grad)
```
